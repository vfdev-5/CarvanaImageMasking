# Example of Affine transformation

```python
import os
import sys

import cv2 

# Project
project_common_path = os.path.dirname('.')
project_common_path = os.path.abspath(os.path.join(project_common_path, '..'))
if not project_common_path in sys.path:
    sys.path.append(project_common_path)
```

```python
from common.image_utils import get_image_data
from tensorpack.utils.utils import fix_rng_seed
from tensorpack.dataflow.imgaug import Affine
```

```python
fix_rng_seed(12345)
```

```python
import matplotlib.pylab as plt
%matplotlib inline
```

```python
img = get_image_data('00087a6bd4dc_14', 'Train')
```

```python
plt.imshow(img)
```

### Scale

```python
aff_tf = Affine(scale=(0.5, 2.0), 
                border=cv2.BORDER_CONSTANT, border_value=0.0)

for _ in range(2):
    m = aff_tf._get_augment_params(img)
    print("Transformation matrix: \n", m)
    timg = aff_tf._augment(img, m)
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.imshow(timg)
    plt.subplot(122)
    plt.imshow(img)
```

### Translation

```python
aff_tf = Affine(translate_frac=(0.2, 0.3), 
                border=cv2.BORDER_CONSTANT, border_value=0.0)

for _ in range(2):
    m = aff_tf._get_augment_params(img)
    print("Transformation matrix: \n", m)
    timg = aff_tf._augment(img, m)
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.imshow(timg)
    plt.subplot(122)
    plt.imshow(img)
```

### Rotation

```python
aff_tf = Affine(rotate_max_deg=45.0, 
                border=cv2.BORDER_CONSTANT, border_value=0.0)

for _ in range(2):
    m = aff_tf._get_augment_params(img)
    print("Transformation matrix: \n", m)
    timg = aff_tf._augment(img, m)
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.imshow(timg)
    plt.subplot(122)
    plt.imshow(img)
```

### Shear

```python
aff_tf = Affine(shear=45.0, 
                border=cv2.BORDER_CONSTANT, border_value=0.0)

for _ in range(2):
    m = aff_tf._get_augment_params(img)
    print("Transformation matrix: \n", m)
    timg = aff_tf._augment(img, m)
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.imshow(timg)
    plt.subplot(122)
    plt.imshow(img)
```

### All together

```python
aff_tf = Affine(scale=(0.5, 2.0), 
                translate_frac=(0.2, 0.3), 
                rotate_max_deg=45.0, 
                shear=45.0, 
                border=cv2.BORDER_CONSTANT, border_value=0.0)

for _ in range(2):
    m = aff_tf._get_augment_params(img)
    print("Transformation matrix: \n", m)
    timg = aff_tf._augment(img, m)
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plt.imshow(timg)
    plt.subplot(122)
    plt.imshow(img)
```
